```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC, SVC
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score, train_test_split, cross_val_predict
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris(as_frame=True)
```

```python
X = iris.data.values
y = (iris.target == 2)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
lsvc_clf = LinearSVC(C = 0.2, random_state=42)
lsvc_clf.fit(X_train, y_train)
predicted = lsvc_clf.predict(X_test)
accuracy_score(y_test, predicted)
```

```python
svc_clf = SVC(C = 1, random_state=42)
svc_clf.fit(X_train, y_train)
predicted = svc_clf.predict(X_test)
accuracy_score(y_test, predicted)
```

```python
sdg_clf = SGDClassifier(random_state = 42)
sdg_clf.fit(X_train, y_train)
predicted = svc_clf.predict(X_test)
accuracy_score(y_test, predicted)
```

```python
from sklearn.pipeline import make_pipeline
lsvc_pipeline = make_pipeline(
    StandardScaler(),
    LinearSVC(C=0.2, random_state = 42)
)
lsvc_pipeline.fit(X_train, y_train)
predicted = lsvc_pipeline.predict(X_test)
accuracy_score(y_test, predicted)
```
