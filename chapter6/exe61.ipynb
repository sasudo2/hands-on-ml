```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
from scipy.stats import uniform
```

```python
tree_clf = DecisionTreeClassifier(random_state=42)
moon = make_moons(n_samples = 10000, noise = 0.4)
train_x, test_x, train_y, test_y = train_test_split(moon[0], moon[1])
tree_clf.fit(train_x, train_y)
tree_clf.score(test_x, test_y)
```

```python
from sklearn.model_selection import cross_val_score
op_tree_clf = DecisionTreeClassifier(random_state = 42)
param_grid = [
    {"max_leaf_nodes": [2, 3, 4, 5, 6],
     "max_depth": [2, 4, 8, 16, 32],
     "criterion": ['gini', 'entropy']}
    ]
grid = GridSearchCV(op_tree_clf, param_grid=param_grid, cv = 3, scoring = "accuracy", n_jobs = -1)
grid.fit(train_x, train_y)
grid.best_estimator_.score(test_x, test_y)
```
